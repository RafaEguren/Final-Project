```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
pd.set_option('display.max_columns', 500)
```

```python
data = pd.read_csv('./vgsales_ml.csv')
data.head()
```

```python
data = data[['Name','Rank','Platform','Year','Genre','Publisher']]
data.head()
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
datadumm = pd.get_dummies(data=data, columns=['Genre'])
datadumm.head()
```

```python
datadumm2 = pd.get_dummies(data=datadumm, columns=['Platform'])
datadumm2.head()
```

```python
datadumm3 = pd.get_dummies(data=datadumm2, columns=['Name'])
datadumm3.head()
```

```python
datadumm4 = pd.get_dummies(data=datadumm3, columns=['Publisher'])
datadumm4.head()
```

```python
data_clusters = kmeans.fit(datadumm4)
data_clusters.cluster_centers_
```

```python
datadumm4['Cluster'] = data_clusters.fit_predict(datadumm4)
```

```python
datadumm4.Cluster.value_counts()
```

```python
datadumm4[datadumm4.Cluster == 2].Year.value_counts()
```

```python
datadumm4.groupby(['Cluster'])['Year'].mean()
```
