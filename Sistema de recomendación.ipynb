```python
import pandas as pd
from sklearn import preprocessing
from scipy.spatial.distance import pdist, squareform
```

```python
data = pd.read_csv('./vgsales_ml.csv')
data1 = pd.read_csv('./rs.csv')
```

```python
data.head()
```

```python
data1.head()
```

```python
data2=data1[['labels']]
data2.head()
```

```python
datars = pd.concat([data, data2], axis=1)
datars.head()
```

```python
datars1 = datars[['Name','Year','NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales','labels']]
datars1.head()
```

```python
datars2 = datars1.set_index('Name').head()
datars2
```

```python
x = datars2.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(x_scaled)
```

```python
df.head()
```

```python
distancias = pd.DataFrame ( 1 / ( 1  + squareform (pdist (df.T, 'euclidean'))), 
                          index = df.columns, columns = df.columns)

distancias
```
