```python
import pandas as pd
from sklearn import preprocessing
from scipy.spatial.distance import pdist, squareform
```

```python
data = pd.read_csv('./vgsales_ml.csv')
data1 = pd.read_csv('./rs.csv')
```

```python
data.head()
```

```python
data1.head()
```

```python
data2=data1[['labels']]
data2.head()
```

```python
datars = pd.concat([data, data2], axis=1)
datars.head()
```

```python
datars1 = datars[['Name','Year','NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales','labels']]
datars1.head()
```

```python
datars2 = datars1.set_index('Name')
datars2.head()
```

```python
x = datars2.values #returns a numpy array
#min_max_scaler = preprocessing.MinMaxScaler()
#x_scaled = min_max_scaler.fit_transform(x)
X_normalized = preprocessing.normalize(x, norm='l2')
df = pd.DataFrame(X_normalized)
```

```python
df.head()
```

```python
columns= ['Year','NA_Sales','EU_Sales','JP_Sales','Other_Sale','Global_Sales','labels']
```

```python
df.columns = columns
```

```python
df.head()
```

```python
datars3=datars[['Name']]
datars3.head()
```

```python
df2 = pd.concat([datars3,df], axis=1)
df2.head()
```

```python
df3 = df2.set_index('Name')
df3.head()
```

```python
df4=df3.T
```

```python
df4.head()
```

```python
distancias = pd.DataFrame ( 1 / ( 1  + squareform (pdist (df4.T, 'euclidean'))), 
                          index = df4.columns, columns = df4.columns)
distancias.head()
```
